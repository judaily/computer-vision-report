```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
# np.random.normal는 정규 분포로부터 난수 생성
# 세 개의 난수 데이터 어레이의 리스트

data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

box = ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5, vert=False)
# vert=False => 수평 방향의 박스 플롯 default 수직 방향
# notch=True => Median(중앙값)의 95% 신뢰 구간의 Notched Boxplot 표시
# 노치의 범위의 IQR(Inter Quartile Range)은 사분위 범위이며, (Q3-Q1)
# whis=2.5로, 수염의 길이가 길어지고 수염 범위 밖 데이터 표시 안됨

# ax.boxplot([data_a, data_b, data_c]) 주어진 데이터 어레이로부터 얻어진 요약 수치를 박스로 표시
ax.set_ylim(-10.0, 10.0)


# 가운데 박스부터 중심선을 Whisker(수염)
# boxplot()는 위쪽 경계부터 Q3 + whis*(Q3-Q1)까지, 아래쪽 경계부터 표시
# whis default 1.5, 수염 외부의 데이터 포인트 (Outlier)는 개별 표시

whiskers = [item.get_ydata() for item in box['whiskers']] # Q1, Q3, max, min 값, 
medians = [item.get_ydata() for item in box['medians']] # 중앙값의 Line2D 인스턴스 리스트
fliers = [item.get_ydata() for item in box['fliers']] # 수염 범위 밖 데이터 포인트 반환
# ax.boxplot()는 박스 플롯의 Line2D 인스턴스 리스트를 딕셔너리로 반환
print('whiskers:', whiskers)
print('medians:', medians)
print('fliers:', fliers)

plt.show()
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4], showmeans=True
                       , quantiles=[[0.25, 0.75], [0.1, 0.9], [0.3, 0.7]])
# ax.violinplot()은 리스트를 딕셔너리(violin)로 반환
# positions는 x 지정, 지정안하면 1,2,3 순서로 표시
# showmeans는 데이터 분포에서 평균값 위치에 직선 표시 Default: False
# showextrema는 최대/최소값에 직선 표시 Default: True
# showmedians는 중간값에 직선 표시 Default: False
# quantiles는 데이터 분포 분위수 직선 표시 0~1값 (ex.25%, 75% 위치)

ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
# ax.set_xticks([1, 2, 3, 4, 5]) # x축 눈금 지정
ax.set_xticklabels(['A', 'B', 'C'])

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')
# bodies 채워진 영역 PolyCollection 인스턴스 리스트.
# set_facecolor()는 n번째 바이올린 분포 영역 색상 지정

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')
# cmeans(평균값), cmins(최소값), cmaxes(최대값), cbars(중심값), cmedians(중간값), 
# cquantiles(분위값) LineCollection 인스턴스

# violin[‘cbars’]의 set_edgecolor()는 중심 직선 색상 지정
# violin[‘cmaxes’]의 set_edgecolor()는 분포의 최대값 직선 색상 지정
# violin[‘cmins’]의 set_edgecolor()는 분포의 최소값 직선 색상 지정
# violin[‘cmeans’]의 set_edgecolor()는 분포의 평균 직선 색상 지정

# 사각형 그리기
ax.add_patch(                       # Patch 추가
   patches.Rectangle(               # 사각형 표시
      (1.8, 1.0),                   # (x, y)
      0.4, 1.5,                     # width, height
      edgecolor = 'deeppink',
      facecolor = 'lightgray',
      fill=True,                    # 도형의 면에 색이 채워짐
   ))

# 화살표 그리기
ax.add_patch(
     patches.Arrow(
        3.0, 5.0, # 화살표 시작점 위치
        0.0, -3, # 위치 변화량 dx, dy
        width=0.3,
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
     ))

# 4. 다각형 그리기
ax.add_patch(
     patches.Polygon(
        ((3.9, 4.0), (4.0, 2.5), (4.1, 4.0)), # 꼭지점 위치
        closed=True,                          # 닫힌 다각형
        edgecolor = 'deeppink',
        facecolor = 'lightgray'
     ))

# 5. 텍스트 삽입하기
ax.text(1.5, 3.5, 'Max of Data A')
ax.text(2.5, 6.0, 'Max of Data B')
ax.text(3.5, 8.0, 'Max of Data C')

plt.show()
```
