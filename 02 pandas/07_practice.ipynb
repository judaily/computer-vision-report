```python
# 열 이름 관리하기
ebola = pd.read_csv('./doit_pandas-master/data/country_timeseries.csv')

print(ebola)
print('---------------')
print(ebola.columns) # 행
print('---------------')
print(ebola.index) # 열
```

```python
print(ebola.iloc[:5, [0,1,2,3,10,11]])
```

```python
ebola_long = pd.melt(ebola, id_vars = ['Date','Day'])

print(ebola_long)
print('-----------------')
print(ebola_long.head())
```

```python
variable_split=ebola_long.variable.str.split('_')

print(variable_split)
print('-----------------')
print(variable_split[:5])
```

```python
print(type(variable_split))
print(type(variable_split[0]))
```

```python
status_values=variable_split.str.get(0)
country_values=variable_split.str.get(1)

print(status_values)
print('1----------------')
print(status_values[:5])
print(' ')
print(status_values[-5:])
print('2----------------')
print(country_values)
print('3----------------')
print(country_values[:5])
print(' ')
print(country_values[-5:])
```

```python
ebola_long.head()
```

```python
# 상태, 국가 분리 각각의 Series
ebola_long['status'] = status_values
ebola_long['country'] = country_values

print(ebola_long)
print('-----------------')
print(ebola_long.head())
```

```python
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split.columns = ['status','country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)

print(variable_split)
print('--------------------')
print(variable_split.columns)
print('--------------------')
print(ebola_parsed)
print('--------------------')
print(ebola_parsed.head())
```

```python
# 여러 열을 하나로 정리하기
weather=pd.read_csv('./doit_pandas-master/data/weather.csv')

print(weather)
print('----------------')
print(weather[:5])
print('----------------')
print(weather.iloc[:5, :])
```

```python
weather_melt=pd.melt(
      weather
    , id_vars=['id','year','month','element'] # 교정 컬럼
    , var_name='day'
    , value_name='temp'
    )

print(weather_melt)
print('-----------------')
print(weather_melt.head())
```

```python
weather_tidy=weather_melt.pivot_table(
    index = ['id','year','month','day'],
    columns='element',
    values='temp',
    # dropna=False
    dropna=True
)

print(weather_tidy)
```

```python
weather_tidy=weather_melt.pivot_table(
    index = ['id','year','month','day'],
    columns='element',
    values='temp',
    dropna=False
)

print(weather_tidy)
```

```python
weather_tidy_flat = weather_tidy.reset_index()

print(weather_tidy_flat)
print('----------------------')
print(weather_tidy_flat.head())
```
